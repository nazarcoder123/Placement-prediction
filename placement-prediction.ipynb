```python
# Now we need to import logistic Regression for that we are using here two indepandent variable x,y
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
# This will return the shape of x how much Row's & Column's it contain
x.shape
```

```python
# This will return the shape of y how much Row's & Column's it contain(In this we have only column's)
y.shape
```

```python
# This is the library we are impoting for training the data,testing and spliting the data
from sklearn.model_selection import train_test_split
```

```python
# In this we are assigning the 10% of the data for testing purposes
train_test_split(x,y,test_size=0.1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
y_test
```

```python
x_test
```

```python
# Now let's do scaling between -1 & +1
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
x_train = Scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test  = Scaler.transform(x_test)
```

```python
x_test
```

```python
# To train the model we have dedicated to call LogisticRegression classifier
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
# Now we need to fit the x_train & y_train data set in this
# Model training
clf.fit(x_train,y_train)
```

```python
# This will display the prediction of the x_test which is done by your model
clf.predict(x_test)
```

```python
# This will give the actual result of the placement data
y_test
```

```python
y_pred = clf.predict(x_test)
```

```python
y_test
```

```python
# To measure the accuracy of the model we import this model
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
